```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
dataset_path= "../dataset/oxford-iiit-pet"
images_path = os.path.join(dataset_path, "images")
masks_path = os.path.join(dataset_path, "annotations/trimaps")
```

```python
image_files = [f for f in os.listdir(images_path) if not f.startswith("._")]
mask_files = [f for f in os.listdir(masks_path) if not f.startswith("._")]

print("Total Images:", len(image_files))
print("Total Masks:", len(mask_files))
```

```python
img_path = os.path.join(images_path, image_files[0])
mask_path = os.path.join(masks_path, mask_files[0])

image = Image.open(img_path)
mask = Image.open(mask_path)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Image")
plt.imshow(image)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Mask")
plt.imshow(mask, cmap="gray")
plt.axis("off")

plt.show()
```

```python
print("Image size:", image.size)
print("Mask size:", mask.size)
```

```python
mask_array = np.array(mask)

print("Unique mask values:", np.unique(mask_array))
```

```python
plt.figure(figsize=(12,6))

for i in range(4):

    img = Image.open(os.path.join(images_path, image_files[i]))
    mask = Image.open(os.path.join(masks_path, mask_files[i]))

    plt.subplot(2,4,i+1)
    plt.imshow(img)
    plt.axis("off")

    plt.subplot(2,4,i+5)
    plt.imshow(mask, cmap="gray")
    plt.axis("off")

plt.show()
```
